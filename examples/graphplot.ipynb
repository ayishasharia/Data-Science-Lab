```python
from matplotlib import pyplot as pt
y=[2,4,6,8,10]
pt.plot(y,'b')
pt.show()
```

```python
import numpy as np
from matplotlib import pyplot as pt
% matplotlib inline
x=np.arange(0,10,1)
y=np.sin(x)
pt.plot(x,y,'y')
pt.show()
pt.bar(x,y)
pt.show()
```

```python

from matplotlib import pyplot as pt
y=[2,4,6]
x=[1,3,5]
pt.plot(x,y,'g')
pt.title('GRAPH')
pt.xlabel('X axis')
pt.ylabel('Y axis')
```

```python
from matplotlib import pyplot as pt
y=[2,4,6]
x=[1,3,5]
pt.plot(x,'r')
pt.plot(y,'b')
pt.title('GRAPH')
pt.xlabel('X axis')
pt.ylabel('Y axis')
pt.legend(['x','y'])
pt.show()
```

```python
from matplotlib import pyplot as pt
y=[1,2,3,4,5,6]
x=[1,2,3,4,5,6]
pt.title('GRAPH')
pt.xlabel('X axis')
pt.ylabel('Y axis')
pt.scatter(x,y)
pt.show()
```
