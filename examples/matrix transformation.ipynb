```python
#rotation
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
from PIL import Image
from scipy import ndimage
a=Image.open("alone.jpeg")
b=np.array(a)
c=ndimage.rotate(b,45,cval=128)
d=Image.fromarray(c)
plt.imshow(d)
```
