```python
import numpy as np
a=np.arange(9).reshape(3,3)
print("first matrix")
print(a)
b=np.arange(9).reshape(3,3)
print("second matrix")
print(b)
c=np.dot(a,b)
print("product od 2 matrcies")
print(c)
t=np.transpose(c)
print("transpose of product matrix")
print(t)
```

```python
#perform matrix transformation using python
#rotation
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
from PIL import Image
from scipy import ndimage
a=Image.open("alone.jpeg")
b=np.array(a)
c=ndimage.rotate(b,45,cval=128)
d=Image.fromarray(c)
plt.imshow(d)
```

```python
#perform matrix transformation using python
#scale
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
from PIL import Image
from scipy import ndimage
a=Image.open("samp.jpg")
b=np.array(a)
c=ndimage.zoom(b,(0.5,0.5,1))
d=Image.fromarray(c)
plt.imshow(d)
```

```python
#perform matrix transformation using python
#shearing
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
from PIL import Image
from scipy import ndimage
a=Image.open("samp.jpg")
b=np.array(a)
h,w,c=b.shape
d=[[1,0,0],[0.5,1,0],[0,0,1]]
shear=ndimage.affine_transformation(b,d)
d=Image.fromarray(shear)
plt.imshow(d)
```
