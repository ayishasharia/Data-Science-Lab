```python
import numpy as np
a=np.array([1,2,3,4])
print("mean:",np.mean(a))
print("median:",np.median(a))
print("variance:",np.var(a))
print("std:",np.std(a))
print("sort:",np.sort(a))
print("sum:",np.sum(a))
print("sin:",np.sin(a))
print("cos:",np.cos(a))



```

```python
from matplotlib import pyplot as pt
import numpy as np
%matplotlib inline
x=np.arange(5,8,2)
y=np.sin(x)
pt.plot(x,y)
pt.show()
pt.bar(x,y)
pt.show()
```

```python
from matplotlib import pyplot as pt
x=[0,1,2,3,4,5,6,7]
y=[1,5,7,9,11,13,15,17]
pt.plot(x,'r')
pt.plot(y,'g*')
pt.xlabel("x axis")
pt.ylabel("y axis")
pt.title("Graph")
pt.legend(['x','y'])
pt.show()
```
