```python
#import
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from sklearn.datasets import load_iris
from sklearn.metrics import mean_squared_error,r2_score
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
#load dataset
irisData=load_iris ()
print("\ndata and targets\n  __________\n")
print(irisData)
#create feature and target arrays
print("\ndata \n __________\n")
print(irisData)
x=irisData.data
print(x)
print("\ntargets\n__________\n")
y=irisData.target
print(y)
#split data to training and testing data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print("\ntrain data\n__________\n")
print(x_train)
print("\ntest data\n__________\n")
print(x_test)
print("\ntrain target\n__________\n")
print(y_train)
print("\ntest target\n_________________\n")
print(y_test)
#traindata+algorthm=model(craete model)
model=LinearRegression()
model.fit(x_train,y_train)                                                                                                                                          
#predict the future
print("\n y intercept\n__________\n")
print(model.intercept_)
#test the accurence
print("\nPredict target and print actual \n__________________________________________\n")
print(model.predict(x_test))
print("\n\n")
print(y_test)

#test the accurance
print("\n accuracy of train data and target\n_____________________________________\n")
print(model.score(x_train,y_train))
print("\n accuracy of test data and target\n__________________________________\n")
print(model.score(x_test,y_test))
print("\n performance of model using mean_square\n_________________________________________\n")
y_predict=model.predict(x_test)
print(mean_squared_error(y_test,y_predict))
print("\n performance of model using r2_score\n______________________________________\n")
print(r2_score(y_test,y_predict))


#graph

print("\n graph \n__________\n")

plt.plot(y_test)
plt.plot(y_predict)
plt.show()
```
